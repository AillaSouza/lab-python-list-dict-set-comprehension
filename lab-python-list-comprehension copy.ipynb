````python
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

def all_prices(products):
    print(" ")
    prices = {product: float(input(f"Enter the price of {product}: ")) for product in products}
    return prices

def get_customer_orders(products):
    num_orders = int(input("Enter the number of customer orders: "))
    
    ordered_products = [input(f"Enter the name of a product that a customer wants to order: ").strip().lower() for _ in range(num_orders)]
    
    customer_orders = {product: ordered_products.count(product) for product in set(ordered_products) if product in products}
    
    invalid_products = set(ordered_products) - set(products)
    if invalid_products:
        print(f"Invalid products: {', '.join(invalid_products)}")
    
    return customer_orders

def order_statistics(customer_orders, products):
    unique_percentage = (len(customer_orders)  / len(products)) * 100 
    
    print(f"\nTotal unique products ordered: {len(customer_orders) }")
    print(f"Percentage of unique products ordered: {unique_percentage:.2f}%")

def update_and_display_inventory(customer_orders, inventory):
    updated_inventory = {product: quantity - customer_orders.get(product, 0) 
                         for product, quantity in inventory.items() if quantity - customer_orders.get(product, 0) > 0}
    
    print("Updated Inventory:")
    for product, quantity in updated_inventory.items():
        print(f"{product}: {quantity} left in stock")
    
    return updated_inventory
    
def total_cost(customer_orders, prices):
    return sum(prices[product] * quantity for product, quantity in customer_orders.items())


````

````python
def main(): 
    products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']
    
    print("Initialize the inventory")
    print(" ")
    inventory = initialize_inventory(products)
    
    print(" ")
    print("Enter the prices for the products")
    prices = all_prices(products) 
    
    print(" ")
    print("Get customer orders")
    customer_orders = get_customer_orders(products)
    
    print(" ")
    order_statistics(customer_orders, products)
    
    print(" ")
    updated_inventory = update_and_display_inventory(customer_orders, inventory)
    
    print(" ")
    print(f"Total cost: €{total_cost(customer_orders, prices):.2f}")
````

````python
if __name__ == "__main__":
    main()
````
